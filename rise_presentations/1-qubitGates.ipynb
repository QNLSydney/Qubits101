# Examples of single qubit gates

Bloch-sphere plots of qubit states before and after applying a single qubit gate 

To plot arbitrary states on the Bloch sphere, we first define the Pauli matrices,

$\hat{\sigma}_z = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}$ $\hat{\sigma}_x = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}$ $\hat{\sigma}_y = \begin{pmatrix} 0 & -i \\ i & 0 \end{pmatrix}$

and a function that returns the expectation value for the state of each one (e.g. $\langle \psi_t | \sigma_z | \psi_t \rangle$)

```python
sigZ = np.array([[1,0],
                 [0,-1]])

sigX = np.array([[0,1],
                 [1,0]])

sigY = np.array([[0,-1j],
                 [1j,0]])

def bsPoints(qState):
    x = (qState.conj().T @ sigX @ qState).real
    y = (qState.conj().T @ sigY @ qState).real
    z = (qState.conj().T @ sigZ @ qState).real
    return [x,y,z]
```

Define an initial state $| \psi(t=0) \rangle = c_0 | E_0 \rangle + c_1 | E_1 \rangle = \begin{pmatrix} c_0 \\ c_1 \end{pmatrix}$ and plot on the Bloch sphere

```python
# psi0 = np.array([1,0])
# psi0 = np.array([1,1j])*1/np.sqrt(2)
# psi0 = np.array([1,-1])*1/np.sqrt(2)
psi0 = np.array([1,np.exp(1j*np.pi/4)])*1/np.sqrt(2)

bsPlot = qutip.Bloch()
bsPlot.add_vectors(bsPoints(psi0))
bsPlot.show()
```

Define the single qubit gate (e.g., $\hat X = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix} = \sigma_x$), apply it to the state and plot again

```python
qGate = sigX

psiF = qGate @ psi0

bsPlot = qutip.Bloch()
bsPlot.add_vectors([bsPoints(psi0),bsPoints(psiF)])
bsPlot.show()
```
